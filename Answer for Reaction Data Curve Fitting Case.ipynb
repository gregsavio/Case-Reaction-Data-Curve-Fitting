```python
import numpy as np
import matplotlib.pyplot as pt
from scipy.optimize import minimize

#data
rd = np.array([0.0192, 0.0234, 0.0258, 0.0281, 0.0318, 0.0319, 0.0334])
pd = np.array([1.382, 2.075, 2.431, 2.762, 3.5, 3.833, 4.129])
#data tambahan
pspan = np.linspace(pd[0],pd[-1],51)
tguess = [0.1,0.1]

#definisi fungsi
def fun(t):
    rs = t[0]*pd/(1+t[1]*pd)
    sse = sum((rs-rd)**2)
    return sse
 
topt = minimize(fun,tguess).x
sse = fun(topt)
rs = topt[0]*pspan/(1+topt[1]*pspan)

print('teta 1 = {:.4f}'.format(topt[0]))
print('teta 2 = {:.4f}'.format(topt[1]))
print('sse = {:.4e}'.format(sse))

pt.plot(pspan,rs,'-',color='black',label='r simulasi')
pt.plot(pd,rd,'o',color='red',label='r data')
pt.title('Hubungan r dan P pada suhu 170C')
pt.xlabel('P, bar')
pt.ylabel('r, mol/s/kg')
pt.legend()
pt.grid()
```
